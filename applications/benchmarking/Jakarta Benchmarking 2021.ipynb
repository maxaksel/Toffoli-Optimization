```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
meas_adjusted_fidelities = (0.8695731996355911, 0.8167916144767129, 0.6914072895001032)
non_meas_adjusted_fidelities = (0.8005898604509247, 0.7546322499197513, 0.6408741660532884)
```

```python
ind = np.arange(3) 
width = 0.35       
plt.bar(ind, meas_adjusted_fidelities, width, label='Measurement Error Mitigation')
plt.bar(ind + width, non_meas_adjusted_fidelities, width,
    label='No Measurement Error Mitigation')

plt.ylabel('Fidelity (%)')
plt.ylim((0.5, 1.0))
plt.title('Fidelity Comparisons of Toffoli Gate Implementations on IBM Hardware')

plt.xticks(ind + width / 2, ('Pulse Optimized', 'Canonical', 'IBM Optimized'))
plt.legend(loc='best')
plt.savefig('fidelities_11_17_high_resolution.png', dpi=400)
plt.show()
```
